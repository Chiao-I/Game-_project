# Knockout

it is a dice game. 2 players
1. to decide who starts each palyer roll a dice. The player with a heighest number satrts the game.
2. There are 10 rounds. Each player rolls two dices. Score is the sum of two dice number.
3. if score is seven, all previous score will be erased.

# Preparation

creating functions:
- welcome(player1_name, player2_name)                       -> welcome players with their names
- compare(player1,player2,player1_score,player2_score)      -> compare two player' results and show who should start the game
- compare_final(player1,player2,player1_score,player2_score)->compare two player' final score
- clean_scores(turns)                                       -> return a list with zeros (the number of zero will eazual the truns of the game)
-  board(player1_score, player2_score, turns, first_player, second_player) -> to visualzed the game board
- throw_a_dice(name)                                        -> instructing players roll a dice and visualize the dice
- throw_two_dice(name)                                      -> instructing players roll a dice and visualize the dice
- replay()                                                  -> asking if players want to play again

```python
#tool to make output colorful

import colorama
from colorama import Fore

```

```python
#welcome function
def welcome(player1,player2):
    print(f"Welcome {player1},{player2}")
    
```

```python
# compare the result to decide who should start the game
def compare(player1,player2,player1_score,player2_score):
        
    if player2_score > player1_score:
        print(Fore.BLACK +'**************************')
        print(Fore.BLACK +f"{player2} start the game")
        print(Fore.BLACK +'**************************')
        return player2
        
    elif player1_score > player2_score:
        print(Fore.BLACK +'**************************')
        print(Fore.BLACK +f"{player1} start the game")
        print(Fore.BLACK +'**************************')
        return player1
     
    elif player1_score == player2_score:
        print(Fore.BLACK +'**************************')
        print(Fore.BLACK +"Role the dice again:")
        print(Fore.BLACK +'**************************')
        
        player1_dice = throw_a_dice(player1)
        print(Fore.BLACK +'*************************************')
        print(Fore.BLACK +f'now {player1} you got {player1_dice}')
        print(Fore.BLACK +'*************************************')

        player2_dice = throw_a_dice(player2)
        print(Fore.BLACK +'*************************************')
        print(Fore.BLACK +f'now {player2} you got {player2_dice}')
        print(Fore.BLACK +'*************************************')

        
```

```python
#compare the final score

def compare_final(player1,player2,player1_score,player2_score):
    
    if sum(player2_score) > sum(player1_score):    
        print(f"{player2} ")
        
        
        print(Fore.MAGENTA +" __      __      __    _____    ____   __    ...")
        print(Fore.MAGENTA +" \ \    /  \    / /     | |     |   \ / |    : :")
        print(Fore.MAGENTA +"  \ \  / /\ \  / /      | |     | |\ V  |    : :")
        print(Fore.MAGENTA +"   \ \/ /  \ \/ /       | |     | | \   |    : :")
        print(Fore.MAGENTA +"    ----    -----      -----    ---  ---     ( )")
        
    elif sum(player2_score) < sum(player1_score): 
        print(f"{player1}")
        
        
        print(Fore.MAGENTA +" __      __      __    _____    ____   __    ...")
        print(Fore.MAGENTA +" \ \    /  \    / /     | |     |   \ / |    : :")
        print(Fore.MAGENTA +"  \ \  / /\ \  / /      | |     | |\ V  |    : :")
        print(Fore.MAGENTA +"   \ \/ /  \ \/ /       | |     | | \   |    : :")
        print(Fore.MAGENTA +"    ----    -----      -----    ---  ---     ( )")

        
     
    elif sum(player2_score) == sum(player1_score):
        print('Draw!')
        


```

```python
#once the score is 7, this function clean all previous scores and return a list with 0

def clean_scores(turns):
    new_list =[]
    while len(new_list) < turns:
        new_list.append(0)
    return new_list
    
```

```python
# to visualize the game board

def board(player1_score, player2_score, turns, first_player, second_player):
    print(f"                {first_player} Vs {second_player}")
    print(" Turn | First Player |  Second Player              ")
    print("--------------------------------------")

    for i, j,t in zip(player1_score, player2_score,range (1, turns+1)):
                      print(f' {t}    |  {i}           |      {j}   ')
    
    player1_total_score = sum(player1_score)
    player2_total_score = sum(player2_score)
    
    
        

    print('--------------------------------------')
    print(f'total |    {player1_total_score}        |       {player2_total_score}          ') 
    print('--------------------------------------')
```

```python
#throw 2 dices

import random
def throw_two_dice(name):
    input(f"Please {name} hit enter to roll dices:")
    dice1 = random.randint(1,6)
    if dice1 == 1:
        print(Fore.RED +" -------")
        print(Fore.RED +"|       |")
        print(Fore.RED +"|   O   |")
        print(Fore.RED +"|       |")
        print(Fore.RED +" -------")
        
    elif dice1 == 2:
        print(Fore.CYAN +" -------")
        print(Fore.CYAN +"| O     |")
        print(Fore.CYAN +"|       |")
        print(Fore.CYAN +"|    O  |")
        print(Fore.CYAN +" -------")
        
    elif dice1 == 3:
        print(Fore.YELLOW +" -------")
        print(Fore.YELLOW +"| O     |")
        print(Fore.YELLOW +"|   O   |")
        print(Fore.YELLOW +"|     O |")
        print(Fore.YELLOW +" -------")
        
    elif dice1 == 4:
        print(Fore.MAGENTA   +" -------")
        print(Fore.MAGENTA   +"| O   o |")
        print(Fore.MAGENTA   +"|       |")
        print(Fore.MAGENTA   +"| o   O |")
        print(Fore.MAGENTA   +" -------")
        
    elif dice1 == 5:
        print(Fore.BLUE +" -------")
        print(Fore.BLUE +"| O   o |")
        print(Fore.BLUE +"|   o   |")
        print(Fore.BLUE +"| o   O |")
        print(Fore.BLUE +" -------")
        
    else:
        print(Fore.GREEN +" -------")
        print(Fore.GREEN +"| O   O |")
        print(Fore.GREEN +"| O   O |")
        print(Fore.GREEN +"| O   O |")
        print(Fore.GREEN +" -------")
    
    
    
    dice2 = random.randint(1,6)
    if dice2 == 1:
        print(Fore.RED +" -------")
        print(Fore.RED +"|       |")
        print(Fore.RED +"|   O   |")
        print(Fore.RED +"|       |")
        print(Fore.RED +" -------")
        
    elif dice2 == 2:
        print(Fore.CYAN +" -------")
        print(Fore.CYAN +"| O     |")
        print(Fore.CYAN +"|       |")
        print(Fore.CYAN +"|    O  |")
        print(Fore.CYAN +" -------")
        
    elif dice2 == 3:
        print(Fore.YELLOW +" -------")
        print(Fore.YELLOW +"| O     |")
        print(Fore.YELLOW +"|   O   |")
        print(Fore.YELLOW +"|     O |")
        print(Fore.YELLOW +" -------")
        
    elif dice2 == 4:
        print(Fore.MAGENTA   +" -------")
        print(Fore.MAGENTA   +"| O   o |")
        print(Fore.MAGENTA   +"|       |")
        print(Fore.MAGENTA   +"| o   O |")
        print(Fore.MAGENTA   +" -------")
        
    elif dice2 == 5:
        print(Fore.BLUE +" -------")
        print(Fore.BLUE +"| O   o |")
        print(Fore.BLUE +"|   o   |")
        print(Fore.BLUE +"| o   O |")
        print(Fore.BLUE +" -------")
        
    else:
        print(Fore.GREEN +" -------")
        print(Fore.GREEN +"| O   O |")
        print(Fore.GREEN +"| O   O |")
        print(Fore.GREEN +"| O   O |")
        print(Fore.GREEN +" -------")
    
    
    
    
    
    
    score = dice1 + dice2
    print(f' {name} got {dice1} and {dice2}')
    print('-!-!-!-!-!-!-!-!-!-!-!-!-!-!-!')
    print(Fore.MAGENTA + "\033[1m" + f'{name}, Your score is  {score}!' + "\033[0m")
    print( '-!-!-!-!-!-!-!-!-!-!-!-!-!-!-!')
    
     
    return score
```

```python
# replay: to ask if players want to play again

def replay():
    answer = input('Do you want to replay? :>>>').lower()
    
    return answer.startswith('y')

        
```

# Game knock_out

```python

#greeting

while True: 
    player1 = input("Hi! Player1, what is your name ? >>>")
    while len(player1)==0:
        player1 = input(Fore.RED + "\033[1m" + "Player 1!, PLEASE ENTER YOUR NAME >>>" + "\033[0m")
    player2 = input("Hi! Player2, what is your name ?  >>>")
    while len(player2)==0:
        player2 = input(Fore.RED + "\033[1m" + "Hi Player 2! PLEASE ENTER YOUR NAME ? >>>"+ "\033[0m")
    welcome(player1, player2)
    
        


    

    turns = 1


#deciding who starts first
    print(Fore.GREEN +" __      __      __ ___    ___    _______             _______  _________  _________  _______   ________")
    print(Fore.GREEN +" \ \    /  \    / / | |____| |   / ____  \           |  _____| |___  ___| | ____  |  | ____ \  |__   __|")
    print(Fore.GREEN +"  \ \  / /\ \  / /  |  ____  |   | |   | |           |_____ \     | |     | |___| |  | |___| |    | |")
    print(Fore.GREEN +"   \ \/ /  \ \/ /   | |    | |   | |___| |            _____\ \    | |     | _____ |  | ___ _/     | |")
    print(Fore.GREEN +"    ----    -----   ---    ---   \_______/           |_______|    |_|     |_|   |_|  |_| \_\      |_|")

    print(Fore.BLACK +"************************************************************")
    print(Fore.BLACK +"Roll one dice, Player with the highest number start the game!")
    print(Fore.BLACK +"************************************************************")
    
    
    player1_dice = throw_a_dice(player1)
    print(f' {player1}, you got {player1_dice}')

    player2_dice = throw_a_dice(player2)
    print(f' {player2}, you got {player2_dice}')

#comparing who got a heiger number, and the player will start
    first_player = compare(player1, player2, player1_dice, player2_dice)
    second_player = [player1, player2]
    
    second_player.remove(first_player)
    second_player = str(second_player[0])
    
    first_player_score = []
    second_player_score = []
    


#game starts-show the board
    board(first_player_score, first_player_score, turns,first_player,second_player)

#first player throw a dice


    while turns <= 10:
        first_player_dice = throw_two_dice(first_player)

        if first_player_dice != 7:
            first_player_score.append(first_player_dice)
        else:
            first_player_score = clean_scores(turns)




        second_player_dice = throw_two_dice(second_player)
    
        if second_player_dice != 7:
            second_player_score.append(second_player_dice)
        else:
            second_player_score = clean_scores(turns)
    



        turns = turns + 1

        board(first_player_score,second_player_score,turns,first_player,second_player)
        
    compare_final(first_player, second_player, first_player_score,second_player_score)
    
    answer = replay()
    
    if answer != True:
        print(Fore.GREEN + " _     _  _______   _____   ______     __(O)__")
        print(Fore.GREEN + "| |   | | | _____|  |_  _|  | ___ \   | ___   |")
        print(Fore.GREEN + "| |___| | | |___      | |   | |  | |  | |___| | ")
        print(Fore.GREEN + "|  ____ | | ____|     | |   | |  | |  |  ___  |")
        print(Fore.GREEN + "| |   | | | |____    _| |   | |__/ |  | |   | |")
        print(Fore.GREEN + "|_|   |_| |______|  \___|   |_____/   |_|   |_|")
        break
```
